```python
!pip install yfinance
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#Definir as ações
acoes = ['META', 'GOOG', 'ORCL', 'MSFT', 'AMZN']
```

```python
#Importar os dados
dados_acoes = yf.download(acoes,period='1d',start='2023-01-01')
```

```python
#Criando o primeiro gráfico
dados_acoes['Close'].plot()
plt.xlabel('Data')
plt.ylabel('Preco de Fechamento')
plt.title('Rastreador de ações')
plt.legend(acoes, fontsize=7)
plt.show()
```

```python
#Normalizando os preços
dados_normalizados = dados_acoes['Close'] / dados_acoes['Close'].iloc[0]
dados_normalizados.plot()
plt.xlabel('Data')
plt.ylabel('Preco Normalizado')
plt.title('Rastreador de ações')
plt.legend(acoes, fontsize=7)
plt.show()
```
